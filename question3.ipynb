```python
# q3_shopping_cart.py

def add_item(item, cart=[]):
    cart.append(item)
    return cart

print("Part A - Bug Demonstration")
print(add_item("apple"))
print(add_item("banana"))
print(add_item("milk", cart=["bread"]))
print(add_item("eggs"))

# Expected Output:
# ['apple']
# ['apple', 'banana']
# ['bread', 'milk']
# ['apple', 'banana', 'eggs']


def add_item_fixed(item, cart=None):
    if cart is None:
        cart = []
    cart.append(item)
    return cart

print("\nPart B - Fixed Function")
print(add_item_fixed("apple"))
print(add_item_fixed("banana"))
print(add_item_fixed("milk", ["bread"]))
print(add_item_fixed("eggs"))

# Create a new shopping cart
def create_cart(owner, discount=0):
    return {
        "owner": owner,
        "items": [],
        "discount": discount
    }


# Add an item to the cart
def add_to_cart(cart, name, price, qty=1):
    item = {
        "name": name,
        "price": price,
        "qty": qty
    }
    cart["items"].append(item)


# Attempt to modify a tuple
def update_price(price_tuple, new_price):
    try:
        # Tuples are immutable.
        # This line raises TypeError.
        price_tuple[0] = new_price
    except TypeError as e:
        print("\nTuple Modification Error:")
        print(e)


# Calculate final total after discount
def calculate_total(cart):
    total = 0

    for item in cart["items"]:
        total += item["price"] * item["qty"]

    discount_amount = total * (cart["discount"] / 100)
    final_total = total - discount_amount

    return final_total

def main():

    # Create two independent carts
    cart1 = create_cart("Alice", 10)
    cart2 = create_cart("Bob", 20)

    # Add items to Alice's cart
    add_to_cart(cart1, "Laptop", 50000, 1)
    add_to_cart(cart1, "Mouse", 800, 2)

    # Add items to Bob's cart
    add_to_cart(cart2, "Keyboard", 1500, 1)
    add_to_cart(cart2, "Monitor", 12000, 2)

    # Display carts
    print("\nCustomer 1 Cart")
    print(cart1)

    print("\nCustomer 2 Cart")
    print(cart2)

    # Calculate totals
    print("\nAlice Total =", calculate_total(cart1))
    print("Bob Total =", calculate_total(cart2))

    # Tuple immutability demonstration
    price = (1000,)

    update_price(price, 1500)


if __name__ == "__main__":
    main()
```
